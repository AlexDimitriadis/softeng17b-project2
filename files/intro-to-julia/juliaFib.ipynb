# Recursive Fibonacci Benchmark - Julia

```julia
using PyPlot
#ioff()

function fib(n)
  if n <= 2
    1
  else
    fib(n-1)+fib(n-2)
  end
end

benchList = [40, 45, 50]
times = []  
for bench in benchList
    tic();
    fib(bench);
    t = toc();
    append!(times, t)
end

bar(benchList,times)
ylabel("Time (sec)")
##gcf()
```
